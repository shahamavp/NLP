```python
import pandas as pd
df1=pd.read_csv('/content/drive/MyDrive/SMS_train.csv',encoding='ISO-8859-1')
df1
```

```python
df2=pd.read_csv('/content/drive/MyDrive/SMS_test.csv',encoding='ISO-8859-1')
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
import seaborn as sns
sns.countplot(x='Label',data=df)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Label']=le.fit_transform(df['Label'])
df
```

```python
df
```

```python
tweets=df.Message_body
tweets
```

```python
tweets=tweets.str.replace('[^a-zA-Z0-9]+','')
tweets
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
tweets=tweets.apply(lambda x:[stemmer.stem(token.lower()) for token in tk.tokenize(x)]).apply(lambda token:' '.join(token))
tweets
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
tweets=tweets.apply(lambda x:[token for token in tk.tokenize(x) if token not in sw]).apply(lambda token:' '.join(token))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(tweets)
print(X)
```

```python
type(X)
```

```python
X=X.toarray()
X
```

```python
vectorizer.get_feature_names_out()
```

```python
y=df['Label'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ab=AdaBoostClassifier(random_state=1)
ab.fit(X_train,y_train)
```

```python
y_pred=ab.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
